```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Calculate Economy Rate
# -----------------------------

economy = df.copy()

# Runs conceded
economy["runs_conceded"] = (
    economy["runs_of_bat"] +
    economy["wide"] +
    economy["noballs"]
)

# Bowling statistics
bowling_stats = economy.groupby("bowler").agg(
    runs_conceded=("runs_conceded", "sum"),
    legal_balls=("wide", lambda x: (
        (economy.loc[x.index, "wide"] == 0) &
        (economy.loc[x.index, "noballs"] == 0)
    ).sum())
).reset_index()

# Overs Bowled
bowling_stats["overs"] = bowling_stats["legal_balls"] / 6

# Economy Rate
bowling_stats["economy"] = (
    bowling_stats["runs_conceded"] /
    bowling_stats["overs"]
)

# Minimum Qualification (20 Overs)
qualified = bowling_stats[
    bowling_stats["legal_balls"] >= 120
]

# Top 10 Economy Bowlers
top10 = qualified.sort_values(
    by="economy",
    ascending=True
).head(10)

# -----------------------------
# Plot
# -----------------------------

plt.figure(figsize=(11,6))

ax = sns.barplot(
    data=top10,
    x="economy",
    y="bowler",
    palette="crest"
)

# Value Labels
for i, value in enumerate(top10["economy"]):
    ax.text(
        value + 0.05,
        i,
        f"{value:.2f}",
        va="center",
        fontsize=11,
        fontweight="bold",
        
    )

# Title
plt.title(
    "Top 10 Economy Bowlers (Minimum 20 Overs)",
    fontsize=18,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Economy Rate",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Bowler",
    fontsize=13,
    fontweight="bold"
)

# Grid
plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

# Remove top/right border
sns.despine()

plt.tight_layout()

plt.show()
```

```python


# --------------------------------------------------
# Calculate Bowling Strike Rate
# --------------------------------------------------

bowling_df = df.copy()

# Valid wickets only (exclude run out & retired hurt)
valid_wickets = [
    "bowled",
    "caught",
    "lbw",
    "stumped",
    "hit wicket"
]

# Runs conceded
bowling_df["runs_conceded"] = (
    bowling_df["runs_of_bat"] +
    bowling_df["wide"] +
    bowling_df["noballs"]
)

# Legal delivery
bowling_df["legal_ball"] = (
    (bowling_df["wide"] == 0) &
    (bowling_df["noballs"] == 0)
).astype(int)

# Wicket credited to bowler
bowling_df["bowler_wicket"] = (
    bowling_df["wicket_type"].isin(valid_wickets)
).astype(int)

# Bowling statistics
strike_rate = bowling_df.groupby("bowler").agg(
    legal_balls=("legal_ball", "sum"),
    wickets=("bowler_wicket", "sum")
).reset_index()

# Remove bowlers with 0 wickets
strike_rate = strike_rate[strike_rate["wickets"] > 0]

# Minimum 20 Overs (120 legal balls)
strike_rate = strike_rate[strike_rate["legal_balls"] >= 120]

# Bowling Strike Rate
strike_rate["bowling_sr"] = (
    strike_rate["legal_balls"] /
    strike_rate["wickets"]
)

# Top 10 (Lowest Strike Rate is Best)
top10_sr = strike_rate.sort_values(
    by="bowling_sr"
).head(10)

# --------------------------------------------------
# Plot
# --------------------------------------------------

plt.figure(figsize=(11,6))

ax = sns.barplot(
    data=top10_sr,
    x="bowling_sr",
    y="bowler",
    palette="rocket_r"
)

# Value Labels
for i, value in enumerate(top10_sr["bowling_sr"]):
    ax.text(
        value + 0.2,
        i,
        f"{value:.1f}",
        va="center",
        fontsize=11,
        fontweight="bold"
    )

plt.title(
    "Best Bowling Strike Rate (Minimum 20 Overs)",
    fontsize=18,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Balls Per Wicket",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Bowler",
    fontsize=13,
    fontweight="bold"
)

plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

sns.despine()

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# -----------------------------
# Valid Wickets
# -----------------------------

valid_dismissals = df[df["wicket_type"].notna()].copy()

def group_wicket_type(wt):
    mapping = {
        "caught": "Caught",
        "bowled": "Bowled",
        "lbw": "LBW",
        "runout": "Run Out",
        "stumped": "Stumped",
        "hit wicket": "Hit Wicket"
    }
    return mapping.get(wt, "Others")

valid_dismissals["wicket_group"] = valid_dismissals["wicket_type"].apply(group_wicket_type)

wicket_dist = valid_dismissals["wicket_group"].value_counts()

order = [
    "Caught",
    "Bowled",
    "Run Out",
    "LBW",
    "Stumped",
    "Hit Wicket",
    "Others"
]

wicket_dist = wicket_dist.reindex(order).dropna()

# Premium Colors
colors = [
    "#233B8B",   # Caught
    "#3F73A8",   # Bowled
    "#F59E0B",   # Run Out
    "#4CAF6A",   # LBW
    "#E53935",   # Stumped
    "#7E57C2",   # Hit Wicket
    "#6C757D"    # Others
]

fig = go.Figure(

    data=[
        go.Pie(
            labels=wicket_dist.index,
            values=wicket_dist.values,
            hole=0.55,
            textinfo="label+percent",
            textfont_size=13,
            marker=dict(
                colors=colors,
                line=dict(color="white", width=3)
            ),
            sort=False
        )
    ]

)

fig.update_layout(

    title=dict(
        text="<b>Wicket Type Distribution — IPL 2026</b>",
        x=0.5,
        font=dict(size=22)
    ),

    annotations=[
        dict(
            text=f"<b>{int(wicket_dist.sum())}</b><br>Total Wickets",
            x=0.5,
            y=0.5,
            showarrow=False,
            font=dict(size=22)
        )
    ],

    legend=dict(
        orientation="v",
        y=0.95,
        x=1.02
    ),

    template="plotly_white",

    width=900,
    height=650,

    margin=dict(
        t=80,
        b=40,
        l=40,
        r=120
    )

)

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------------
# Dot Ball Specialists
# -----------------------------------------

dot_df = df.copy()

# Legal dot ball
dot_df["dot_ball"] = (
    (dot_df["runs_of_bat"] == 0) &
    (dot_df["wide"] == 0) &
    (dot_df["noballs"] == 0)
).astype(int)

# Legal delivery
dot_df["legal_ball"] = (
    (dot_df["wide"] == 0) &
    (dot_df["noballs"] == 0)
).astype(int)

# Bowling Summary
dot_stats = dot_df.groupby("bowler").agg(
    dot_balls=("dot_ball","sum"),
    legal_balls=("legal_ball","sum")
).reset_index()

# Overs
dot_stats["overs"] = dot_stats["legal_balls"] / 6

# Minimum Qualification (20 Overs)
dot_stats = dot_stats[dot_stats["legal_balls"] >= 120]

# Top 10
top10_dot = dot_stats.sort_values(
    by="dot_balls",
    ascending=False
).head(10)

# -----------------------------------------
# Plot
# -----------------------------------------

plt.figure(figsize=(11,6))

ax = sns.barplot(
    data=top10_dot,
    x="dot_balls",
    y="bowler",
    palette="crest"
)

# Value Labels
for i, value in enumerate(top10_dot["dot_balls"]):
    ax.text(
        value + 1,
        i,
        f"{int(value)}",
        va="center",
        fontsize=11,
        fontweight="bold"
    )

plt.title(
    "Top 10 Dot Ball Specialists (Minimum 20 Overs)",
    fontsize=18,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Dot Balls Bowled",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Bowler",
    fontsize=13,
    fontweight="bold"
)

plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

sns.despine()

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------
# Total Runs Per Innings
# ---------------------------------------------

df["total_runs"] = df["runs_of_bat"] + df["extras"]

innings_total = (
    df.groupby(["match_no", "innings", "batting_team"], as_index=False)
      .agg(Total_Runs=("total_runs", "sum"))
)

# ---------------------------------------------
# Average Team Score
# ---------------------------------------------

avg_score = (
    innings_total
    .groupby("batting_team", as_index=False)
    .agg(Average_Score=("Total_Runs", "mean"))
    .sort_values("Average_Score", ascending=False)
)


plt.figure(figsize=(11,6))

ax = sns.barplot(
    data=avg_score,
    x="Average_Score",
    y="batting_team",
    hue="batting_team",
    palette="viridis_r",
    dodge=False,
    legend=False
)

# Value Labels
for i, value in enumerate(avg_score["Average_Score"]):
    ax.text(
        value + 0.5,
        i,
        f"{value:.1f}",
        va="center",
        fontsize=11,
        fontweight="bold",
        color="black"
    )

# Title
plt.title(
    "Average Team Score Per Innings — IPL 2026",
    fontsize=18,
    fontweight="bold",
    pad=15
)

# Labels
plt.xlabel(
    "Average Runs",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Team",
    fontsize=13,
    fontweight="bold"
)

# Grid
plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

sns.despine()

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# -------------------------------
# Team Boundary Comparison
# -------------------------------

team_boundary = (
    df.groupby("batting_team")
      .agg(
          Fours=("runs_of_bat", lambda x: (x == 4).sum()),
          Sixes=("runs_of_bat", lambda x: (x == 6).sum())
      )
      .reset_index()
)

team_boundary["Total_Boundaries"] = (
    team_boundary["Fours"] + team_boundary["Sixes"]
)

team_boundary = team_boundary.sort_values(
    "Total_Boundaries",
    ascending=False
)

# -------------------------------
# Official IPL Colors
# -------------------------------

team_colors = {
    "CSK": "#F9CD05",
    "MI": "#004BA0",
    "RCB": "#D71920",
    "KKR": "#3A225D",
    "SRH": "#F26522",
    "DC": "#17449B",
    "PBKS": "#ED1B24",
    "RR": "#EA1A85",
    "GT": "#1C1C1C",
    "LSG": "#00AEEF"
}

# X locations
x = np.arange(len(team_boundary))
width = 0.36

plt.figure(figsize=(12,6))

# Fours
plt.bar(
    x-width/2,
    team_boundary["Fours"],
    width,
    color= "#2563EB",
    label="Fours"
)

# Sixes
plt.bar(
    x+width/2,
    team_boundary["Sixes"],
    width,
    color="#F59E0B",
    label="Sixes"
)

# Value labels
for i, value in enumerate(team_boundary["Fours"]):
    plt.text(
        i-width/2,
        value+3,
        str(value),
        ha="center",
        fontsize=10,
        fontweight="bold"
    )

for i, value in enumerate(team_boundary["Sixes"]):
    plt.text(
        i+width/2,
        value+3,
        str(value),
        ha="center",
        fontsize=10,
        fontweight="bold"
    )

plt.xticks(
    x,
    team_boundary["batting_team"],
    fontsize=12,
    fontweight="bold"
)

# Team names in official colors
ax = plt.gca()

for tick in ax.get_xticklabels():
    team = tick.get_text()
    if team in team_colors:
        tick.set_color(team_colors[team])

plt.title(
    "Team Boundary Comparison — IPL 2026",
    fontsize=20,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Teams",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Boundary Count",
    fontsize=13,
    fontweight="bold"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.35
)

plt.legend(
    fontsize=11,
    frameon=False
)

sns.despine()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# ------------------------------------
# Total Runs Per Ball
# ------------------------------------

df["total_runs"] = df["runs_of_bat"] + df["extras"]

# ------------------------------------
# Over Number (0.1 -> 1, 5.6 -> 6)
# ------------------------------------

df["over_number"] = (
    df["over"]
    .astype(str)
    .str.split(".")
    .str[0]
    .astype(int)
    + 1
)

# ------------------------------------
# Assign Match Phase
# ------------------------------------

def get_phase(over):
    if over <= 6:
        return "Powerplay"
    elif over <= 15:
        return "Middle Overs"
    else:
        return "Death Overs"

df["phase"] = df["over_number"].apply(get_phase)

# ------------------------------------
# Runs in each innings for every phase
# ------------------------------------

phase_runs = (
    df.groupby(
        ["match_no", "innings", "batting_team", "phase"],
        as_index=False
    )
    .agg(Runs=("total_runs", "sum"))
)

# ------------------------------------
# Average phase runs by team
# ------------------------------------

team_phase = (
    phase_runs.groupby(
        ["batting_team", "phase"],
        as_index=False
    )
    .agg(Average_Runs=("Runs", "mean"))
)

# ------------------------------------
# Plotly Graph
# ------------------------------------

fig = px.bar(
    team_phase,
    x="batting_team",
    y="Average_Runs",
    color="phase",
    barmode="group",
    text_auto=".1f",
    color_discrete_map={
        "Powerplay": "#1E88E5",
        "Middle Overs": "#43A047",
        "Death Overs": "#E53935"
    }
)

fig.update_traces(textposition="outside")

fig.update_layout(
    title="<b>Average Runs by Match Phase — IPL 2026</b>",
    template="plotly_white",
    width=1000,
    height=600,
    xaxis_title="<b>Team</b>",
    yaxis_title="<b>Average Runs</b>",
    legend_title="<b>Phase</b>",
    font=dict(size=14)
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# ---------------------------------------------------
# Total Runs
# ---------------------------------------------------

df["total_runs"] = df["runs_of_bat"] + df["extras"]

# ---------------------------------------------------
# Legal Balls (Wide & No Ball are NOT legal deliveries)
# ---------------------------------------------------

team_rr = (
    df.groupby("batting_team")
      .apply(lambda x: pd.Series({
          "Total_Runs": x["total_runs"].sum(),
          "Legal_Balls": ((x["wide"] == 0) & (x["noballs"] == 0)).sum()
      }))
      .reset_index()
)

# Overs
team_rr["Overs"] = team_rr["Legal_Balls"] / 6

# Run Rate
team_rr["Run_Rate"] = (
    team_rr["Total_Runs"] / team_rr["Overs"]
).round(2)

# Sort
team_rr = team_rr.sort_values("Run_Rate", ascending=False)

# ---------------------------------------------------
# Official IPL Team Colors
# ---------------------------------------------------

team_colors = {
    "CSK": "#F9CD05",
    "MI": "#004BA0",
    "RCB": "#D71920",
    "KKR": "#3A225D",
    "RR": "#EA1A85",
    "SRH": "#F26522",
    "PBKS": "#ED1B24",
    "DC": "#17479E",
    "LSG": "#00AEEF",
    "GT": "#1C1C1C"
}

# ---------------------------------------------------
# Plotly Line Chart
# ---------------------------------------------------

fig = px.line(
    team_rr,
    x="batting_team",
    y="Run_Rate",
    markers=True
)

# Thick Line
fig.update_traces(
    line=dict(color="#2E86DE", width=5),
    marker=dict(
        size=15,
        color=[team_colors[t] for t in team_rr["batting_team"]],
        line=dict(color="white", width=2)
    ),
    text=team_rr["Run_Rate"],
    textposition="top center"
)

fig.update_layout(
    title=dict(
        text="<b>Team Run Rate Comparison — IPL 2026</b>",
        x=0.5,
        font=dict(size=24)
    ),
    template="plotly_white",
    width=950,
    height=550,

    xaxis=dict(
        title="<b>Teams</b>",
        tickfont=dict(size=13)
    ),

    yaxis=dict(
        title="<b>Run Rate</b>",
        tickfont=dict(size=13)
    ),

    hovermode="x unified",
    showlegend=False
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# -----------------------------------------
# Extras by Bowling Teams
# -----------------------------------------

extras_df = (
    df.groupby("bowling_team")
      .agg(
          Wides=("wide", "sum"),
          No_Balls=("noballs", "sum"),
          Byes=("byes", "sum"),
          Leg_Byes=("legbyes", "sum")
      )
      .reset_index()
)

extras_df["Total"] = (
    extras_df["Wides"] +
    extras_df["No_Balls"] +
    extras_df["Byes"] +
    extras_df["Leg_Byes"]
)

# Sort by Total Extras
extras_df = extras_df.sort_values("Total", ascending=False)

# -----------------------------------------
# Plot
# -----------------------------------------

fig = go.Figure()

# Wides
fig.add_bar(
    x=extras_df["bowling_team"],
    y=extras_df["Wides"],
    name="Wides",
    marker_color="#EF4444",
    hovertemplate="<b>%{x}</b><br>Wides: %{y}<extra></extra>"
)

# No Balls
fig.add_bar(
    x=extras_df["bowling_team"],
    y=extras_df["No_Balls"],
    name="No Balls",
    marker_color="#F59E0B",
    hovertemplate="<b>%{x}</b><br>No Balls: %{y}<extra></extra>"
)

# Byes
fig.add_bar(
    x=extras_df["bowling_team"],
    y=extras_df["Byes"],
    name="Byes",
    marker_color="#3B82F6",
    hovertemplate="<b>%{x}</b><br>Byes: %{y}<extra></extra>"
)

# Leg Byes
fig.add_bar(
    x=extras_df["bowling_team"],
    y=extras_df["Leg_Byes"],
    name="Leg Byes",
    marker_color="#22C55E",
    hovertemplate="<b>%{x}</b><br>Leg Byes: %{y}<extra></extra>"
)

# -----------------------------------------
# Total Labels
# -----------------------------------------

fig.add_trace(
    go.Scatter(
        x=extras_df["bowling_team"],
        y=extras_df["Total"] + 3,
        mode="text",
        text=extras_df["Total"],
        textfont=dict(
            size=15,
            color="black",
            family="Arial Black"
        ),
        showlegend=False,
        hoverinfo="skip"
    )
)

# -----------------------------------------
# Layout
# -----------------------------------------

fig.update_layout(

    barmode="stack",

    title=dict(
        text="<b>Extras Conceded by Bowling Teams — IPL 2026</b>",
        x=0.5,
        font=dict(size=24)
    ),

    template="plotly_white",

    width=1050,
    height=620,

    xaxis=dict(
        title="<b>Bowling Team</b>",
        tickfont=dict(
            size=13,
            family="Arial Black"
        )
    ),

    yaxis=dict(
        title="<b>Total Extras</b>",
        showgrid=True,
        gridcolor="rgba(180,180,180,0.25)",
        gridwidth=1,
        range=[0, extras_df["Total"].max()+20]
    ),

    legend=dict(
        orientation="h",
        y=1.10,
        x=0.5,
        xanchor="center",
        title="<b>Extra Type</b>",
        font=dict(size=13)
    ),

    hovermode="x unified",

    margin=dict(
        t=110,
        b=60,
        l=60,
        r=40
    )
)

# -----------------------------------------
# Unified Hover
# -----------------------------------------

fig.update_traces(
    hovertemplate="<b>%{x}</b><br>%{fullData.name}: %{y}<extra></extra>"
)

fig.show()
```
